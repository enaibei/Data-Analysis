### Visualisation with Pandas

```python
population_raw = pd.read_csv('population_total.csv')
```

```python
population_raw
```

```python
population_raw.dropna(inplace=True) # replaces old dataframe
```

```python
df_pivot = population_raw.pivot(index="year", columns='country', values='population')
```

```python
df_pivot
```

```python
df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
df_pivot
```

```python
# Lineplot
df_pivot.plot(kind='line',)
```

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', 
              title='Popolation (1955-2020)',
              figsize=(8,4))

import matplotlib.pyplot as plt


plt.savefig('my_test.png')
plt.show
```

```python
# Barplot
df_pivot
```

```python
df_pivot.index
```

```python
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
df_pivot_2020 = df_pivot_2020.T #transpose
```

```python
df_pivot_2020
```

```python
df_pivot_2020.plot(kind='bar')
```

```python
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Year', ylabel='Population',
                  title='Population 2020')
```

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_sample
```

```python
df_pivot_sample.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population Sample')
```
